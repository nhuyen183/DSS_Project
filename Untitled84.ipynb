#1) Hill clibimbing 


```python
def sum_nutrition(df_cal):
  sumProtein = np.sum(df_cal['Protein'])
  sumCarbon = np.sum(df_cal['Carbon'])
  sumFat = np.sum(df_cal['Fat'])
  sumCalo = np.sum(df_cal['calories'])
  return sumProtein, sumCarbon,sumFat
```

```python

protein, carbon, fat  = 500, 500, 200 
def hill_climbing(fat,carbon, protein):
  sumProtein,sumFat, sumCarbon = 0,0,0
  df_cal = df_new[df_new['dishTypes']=='lunch,main course,main dish,dinner'].sample(n=2,replace=True)
  
  a,b,c = sum_nutrition(df_cal)
  if ((a < protein) and (b < carbon) and (c < fat)):
    while ((a < protein) and (b < carbon) and (c < fat)):
      new_row = df_new[df_new['dishTypes']!='lunch,main course,main dish,dinner'].sample(n=1,replace=True)  
      df_cal = df_cal.append(new_row, ignore_index= False)
      a,b,c = sum_nutrition(df_cal)
    print(a,b,c)
    return df_cal
    
  
print("benchmark :",fat, carbon, protein)
hill_climbing( protein,carbon,  fat)


```
