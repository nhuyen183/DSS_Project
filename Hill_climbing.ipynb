```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv('Clean_food.csv')
df.head()
```

```python
fat = df.weightPerServing * df['percentFat']/100
Carbon = df.weightPerServing * df['percentCarbs']/100
Protein = df.weightPerServing * df['percentProtein']/100
```

```python

df = df[['id', 'title','weightPerServing','dishTypes','calories']]
df['Fat']= fat
df['Carbon']= Carbon
df['Protein']= Protein
df.head()
```

#1) Hill clibimbing 


```python
def sum_nutrition(df_cal):
  sumProtein = np.sum(df_cal['Protein'])
  sumCarbon = np.sum(df_cal['Carbon'])
  sumFat = np.sum(df_cal['Fat'])
  return sumProtein, sumCarbon,sumFat
```

```python
def auto_gen_meal( N, df):
  df_cal = df[df['dishTypes']=='lunch,main course,main dish,dinner'].sample(n=N,replace=True)
  return df_cal
```

```python
def meal_eval(df1,p,ca,f):
  a,b,c = sum_nutrition(df1)
  eval = np.abs(a-f) + np.abs(b-ca) + np.abs(c-p)
  return eval
```

```python
def reset_basket(dfs):
  dfs = pd.DataFrame(columns=df.columns)
  return dfs
```

```python
def fill_meal(df,protein,carbon,fat ):
  iter=0
  a,b,c = 0,0,0
  while (( a < protein) or (b < carbon) or (c < fat)):
      df_cal = auto_gen_meal(2,df)
      new_row = df[df['dishTypes']!='lunch,main course,main dish,dinner'].sample(n=1,replace=True)  
      df_cal = df_cal.append(new_row, ignore_index= False)
      iter+=1
      if iter > 16: 
        reset_basket(df)
        iter = 0
      a,b,c = sum_nutrition(df_cal)
  return df_cal
```

```python

def hill_climbing(protein,carbon, fat): 
  optimal = fill_meal(df, protein, carbon, fat)
  iter =0
  for i in range(0,10):
    df1 = fill_meal(df, protein, carbon, fat)
    if meal_eval(df1,protein, carbon, fat) <= meal_eval(optimal,protein, carbon, fat):
      optimal = df1
    print(iter)
    iter+=1
  p,ca,f = sum_nutrition(optimal) 
  print(p,ca,f)
  return optimal
```
